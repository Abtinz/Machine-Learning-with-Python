```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
%reload_ext autoreload
%autoreload 2
```

```python
#simple plot
X = np.linspace(0,5,10)
print(f"X:\n{X}")
Y = X**2
plt.plot(X,Y)
plt.title("Squared Chart")
plt.xlabel("X")
plt.ylabel("y")
```

```python
#subplot and colors
plt.subplot(1,2,1)
plt.plot(X,Y,"r")
plt.title("First Squared Chart")
plt.subplot(1,2,2)
plt.plot(X,Y,"blue")
plt.title("Scond Squared Chart")
```

```python
#figure and plot lable
plotFigure = plt.figure(figsize=(5,5))
axes = plotFigure.add_axes([1,1,1,1])
axes.set_title("Inverse function")
#lable on the plot
axes.plot(X,Y,"black",label = "X-X^2")
axes.plot(Y,X,"r",label = "X^2-X")
axes.legend(loc=0) # locating lable
```

```python
fig , axeses = plt.subplots(figsize = (8,4) , ncols = 3 , nrows = 1)
#seperat plots
plt.tight_layout()
#plots with index
axeses[0].set_title("First Plot")
axeses[0].plot(X,Y,"black")
axeses[1].set_title("Second Plot")
axeses[1].plot(X,Y*X,"red")
axeses[2].set_title("third Plot")
axeses[2].plot(X,X,"green")
```
