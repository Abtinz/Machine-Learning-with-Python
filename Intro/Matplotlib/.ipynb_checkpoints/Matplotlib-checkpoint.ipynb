```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
%reload_ext autoreload
%autoreload 2
```

```python
#simple plot
X = np.linspace(0,5,10)
print(f"X:\n{X}")
Y = X**2
plt.plot(X,Y)
plt.title("Squared Chart")
plt.xlabel("X")
plt.ylabel("y")
```

```python
#subplot and colors
plt.subplot(1,2,1)
plt.plot(X,Y,"r")
plt.title("First Squared Chart")
plt.subplot(1,2,2)
plt.plot(X,Y,"blue")
plt.title("Scond Squared Chart")
```

```python
#figure and plot lable
plotFigure = plt.figure(figsize=(5,5))
axes = plotFigure.add_axes([1,1,1,1])
axes.set_title("Inverse function")
#lable on the plot
axes.plot(X,Y,"black",label = "X-X^2")
axes.plot(Y,X,"r",label = "X^2-X")
axes.legend(loc=0) # locating lable
```

```python
fig , axeses = plt.subplots(figsize = (8,4) , ncols = 3 , nrows = 1)
#seperat plots
plt.tight_layout()
#plots with index
axeses[0].set_title("First Plot")
axeses[0].plot(X,Y,"black")
axeses[1].set_title("Second Plot")
axeses[1].plot(X,Y*X,"red")
axeses[2].set_title("third Plot")
axeses[2].plot(X,X,"green")
```

```python
df = pd.read_csv("C:/Users/ABTIN/Documents/coding/Machine Learning/CSV_FILES/china_gdp.csv" , encoding = "ISO-8859-1",low_memory=False)
df.tail()
```

```python
Xinputs = df[["Year"]]
Yinputs = df[["Value"]]
print(f"X shape and counts :{Xinputs.shape}")    
```

```python
plotFigure = plt.figure(figsize=(5,5))
chine_gdp_axes = plotFigure.add_axes([1,1,1,1])
chine_gdp_axes.set_title("China GDP")
chine_gdp_axes.plot(Xinputs,Yinputs)
chine_gdp_axes.annotate("2010 Gdp" , xy = (2010,6.039659e+12), xytext = (1990,4.039659e+12),arrowprops = dict(facecolor = "black"))
```
