```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,10,1.0)
phiz = 1.0 / (1.0 + np.exp(-x))
```

```python
plt.plot(x,phiz)
plt.title("logestic function plot");
```

```python
plt.plot(x,phiz,'r')
plt.axvline(0, ls = 'dotted' , color = 'b')
plt.axhline(0.5, ls = 'dotted' , color = 'b')
plt.xlabel('X')
plt.ylabel('$\phi (x)$')
plt.title("logestic function plot");
```
