```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
IQ_data = np.array([13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30,33, 33, 35, 35, 35, 36,
40, 45, 46, 52, 70])
IQ_data.sort()
print(IQ_data)
```

```python
'''
    This functi
'''
def box_plot(data):
    
    Q1 = np.quantile(data , 0.25)
    Q2 = np.quantile(data , 0.5)
    Q3 = np.quantile(data , 0.75)
    print("first quantile: ",Q1)
    print("second quantile: ",Q2)
    print("third quantile: ",Q3)

    iqr = Q3 - Q1
    print("IQR (Q3 - Q1) : ", iqr)

    LV = Q1 - (1.5 * iqr)
    print("LV: ",LV)
    
    plt.boxplot(IQ_data, patch_artist=True)
    plt.title("IQ BOX PLOT")
    plt.ylabel("IQ Rate")
```

```python
box_plot(IQ_data)
```
