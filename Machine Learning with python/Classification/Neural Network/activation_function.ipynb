```python
def sigmoid(func):
    return 1/(1+np.exp(-func))
```

```python
x = np.linspace(-50,50,1000)
```

```python
plt.plot(x,sigmoid(x))
plt.grid()
plt.title("sigmoid function")
plt.text(-48,0.5,r'$\sigma(x)=\frac{1}{1+e^{-x}}$',fontsize = 20);
```

# RELU

```python
def ReLU(x):
    return np.maximum(0.0,x)
```

```python
plt.plot(x,ReLU(x),'black')
plt.title("Relu function")
plt.text(-48,10,r'Relu',fontsize = 20,color = "green");
```
