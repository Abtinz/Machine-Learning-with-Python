# Preprocessing

```python
def split_name(name):
    if '.' in name : return name.split(',')[1].split('.')[0].strip()
    else : return 'none'
```

```python
def finding_name_title(splited_string):
     if 'Mr' in splited_string : return 1
     elif 'Master' in splited_string : return 2
     elif ('Ms' in splited_string) or  ('Mlle' in splited_string)  or ('Miss' in splited_string) : return 3
     elif ('Mme' in splited_string) or  ('Mrs' in splited_string)  : return 4
     else: return 5
```

```python
df['Title'] = df['Name'].apply(split_name).apply(finding_name_title)
df.Title.describe()
```

```python
test_df['Title'] = test_df['Name'].apply(split_name).apply(finding_name_title)
test_df.Title.describe()
```

```python
df = df.drop(['PassengerId', 'Name' , 'Ticket' , 'Cabin'],axis = 1)
dummised_df = pd.get_dummies(df.Embarked)

df = df.join(dummised_df)
df = df.drop(['Embarked'],axis = 1)
df.head()
```

```python
test_df = test_df.drop(['PassengerId', 'Name' , 'Ticket' , 'Cabin'],axis = 1)
dummised_test_df = pd.get_dummies(test_df.Embarked)

test_df = test_df.join(dummised_test_df)
test_df = test_df.drop(['Embarked'],axis = 1)
test_df.head()
```

```python
test_df.Fare.fillna(test_df.Fare.median(),inplace = True)
```

```python
df.Age = df.groupby(['Pclass']).Age.transform(lambda x:x.fillna(x.mean())) 
```

```python
test_df.Age = test_df.groupby(['Pclass']).Age.transform(lambda x:x.fillna(x.mean())) 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
df['Sex'] = label.fit_transform(df['Sex'])
df['Sex'].describe()
```

```python
label = LabelEncoder()
test_df['Sex'] = label.fit_transform(test_df['Sex'])
test_df['Sex'].describe()
```

```python
model = LogisticRegression()
```
