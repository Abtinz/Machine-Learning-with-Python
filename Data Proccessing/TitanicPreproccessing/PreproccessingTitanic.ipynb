```python
def split_name(name):
    if '.' in name : return name.split(',')[1].split('.')[0].strip()
    else : return 'none'
```

```python
def finding_name_title(splited_string):
     if 'Mr' in splited_string : return 1
     elif 'Master' in splited_string : return 2
     elif ('Ms' in splited_string) or  ('Mlle' in splited_string)  or ('Miss' in splited_string) : return 3
     elif ('Mme' in splited_string) or  ('Mrs' in splited_string)  : return 4
     else: return 5
```

```python
df['Title'] = df['Name'].apply(split_name).apply(finding_name_title)
df.Title.describe()
```

```python
df.head(15)
```

```python
crosstabed_df = pd.crosstab(df.Title , df.Survived)
crosstabed_df.head()
```

```python
percented_crosstabed_df = crosstabed_df.div(crosstabed_df.sum(1).astype(float) , axis = 0)
percented_crosstabed_df
```

```python
def pd_bar_plot(plotting_df,title,Xlabel,Ylabel):
    plotting_df.plot(kind = 'bar',stacked = True)
    plt.title(title)
    plt.xlabel(Xlabel)
    plt.ylabel(Ylabel)
    plt.show()
```

```python
pd_bar_plot(percented_crosstabed_df,'Survived Rate By Name Title','Survived Rate','Title')
```
